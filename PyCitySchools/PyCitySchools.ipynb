```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = len(student_data["school_name"].unique())
#Calculate the total number of students
students_number = student_data["student_name"].count()
#Calculate the total budget
total_budget = school_data["budget"].sum()
#Calculate the average maths score
average_maths_score = student_data["maths_score"].mean()
#Calculate the average reading score
average_reading_score = student_data["reading_score"].mean()
#Calculate the percentage of students with a passing maths score (50 or greater
passing_maths = student_data.loc[(student_data["maths_score"]>=50)]
percentage_passing_maths = (len(passing_maths))/(students_number)*100
#Calculate the percentage of students with a passing reading score (50 or greater)
passing_reading = student_data.loc[(student_data["reading_score"]>=50)]
percentage_passing_reading = (len(passing_reading))/(students_number)*100
#Calculate the percentage of students who passed maths and reading (% Overall Passing)
passing_both = student_data.loc[(student_data["reading_score"]>=50) & (student_data["maths_score"]>=50)]
percentage_passing_both = (len(passing_both))/(students_number)*100
#Create a dataframe to hold the above results
Area_summary = [{"Total_schools": total_schools,"Total_students": students_number,"Total budget":total_budget,
                           "Average_maths_score": average_maths_score,"Average_reading_score": average_reading_score,
                          "% passing maths":percentage_passing_maths,"% passing reading":percentage_passing_reading,
                          "% overall passing":percentage_passing_both}]
Local_area_government_summary_df = pd.DataFrame(Area_summary)
Local_area_government_summary_df
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#creating the base data frame for the school summary using groupby on the school_data_complete data frame
school_data_df= school_data_complete.groupby(["school_name","type"]).agg({"Student ID":'count',
                                                        "budget":"mean",
                                                        "maths_score":"mean",
                                                        "reading_score":"mean"})
#renaming the columns
school_data_df.rename(columns = {"Student ID": "Total Students","budget":"Total School Budget",
                                 "maths_score":"Average Maths Score","reading_score":"Average Reading Score"},inplace = True)
#calculating per student budget
school_data_df["Total School Budget"].astype("float")
per_student_budget = school_data_df["Total School Budget"]/school_data_df["Total Students"]
school_data_df["Per Student Budget"]= per_student_budget
###determing the % passing maths###
#determining the rows in the school data complete that contain the maths score eqaul or greater than 50
maths=school_data_complete.loc[(school_data_complete["maths_score"]>=50),:]
#counting the students per school with score above or equal to 50
pass_maths=maths.groupby("school_name")["maths_score"].count()
#determining the percentage pass in maths per school
school_data_df["% Passing Maths"]= (pass_maths / school_data_df["Total Students"])*100
#determining the % reading using the same steps as for maths 
reading = school_data_complete.loc[(school_data_complete["reading_score"]>=50),:]
pass_reading=reading.groupby("school_name")["reading_score"].count()
school_data_df["% Passing Reading"]= (pass_reading / school_data_df["Total Students"])*100
#determining the % overall passing using the same steps as above for maths
overall= school_data_complete.loc[(school_data_complete["reading_score"]>=50)&(school_data_complete["maths_score"]>=50),:]
pass_overall = overall.groupby("school_name")["reading_score"].count()
school_data_df["% Overall Passing"]= (pass_overall/ school_data_df["Total Students"])*100
#reseting the index("type") back to column
school_data_df_reset=school_data_df.reset_index("type")
#renaming the type column
school_data_df_reset.rename(columns ={"type":"School Type"},inplace = True)
#rearranging the columns
school_summary_df= school_data_df_reset[["School Type","Total Students","Total School Budget",
                                         "Per Student Budget","Average Maths Score","Average Reading Score",
                                         "% Passing Maths","% Passing Reading","% Overall Passing"]]
#rounding the data to two decimal places
school_summary_df["Average Maths Score"]=school_summary_df["Average Maths Score"].round(2)
school_summary_df["Average Reading Score"]=school_summary_df["Average Reading Score"].round(2)
school_summary_df["% Passing Maths"]=school_summary_df["% Passing Maths"].round(2)
school_summary_df["% Passing Reading"]=school_summary_df["% Passing Reading"].round(2)
school_summary_df["% Overall Passing"]=school_summary_df["% Overall Passing"].round(2)
school_summary_df


```

## Top Performing Schools (By % Overall Passing)

```python
Top_Perfoming_Schools_df = school_summary_df.sort_values("% Overall Passing", ascending = False)
Top_Perfoming_Schools_df.head()
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
year_9_maths_df=school_data_complete.loc[(school_data_complete["year"]==9),:]
year_9_maths_series = year_9_maths_df.groupby("school_name")["maths_score"].mean()
year_10_maths_df=school_data_complete.loc[(school_data_complete["year"]==10),:]
year_10_maths_series = year_10_maths_df.groupby("school_name")["maths_score"].mean()
year_11_maths_df=school_data_complete.loc[(school_data_complete["year"]==11),:]
year_11_maths_series = year_11_maths_df.groupby("school_name")["maths_score"].mean()
year_12_maths_df=school_data_complete.loc[(school_data_complete["year"]==12),:]
year_12_maths_series = year_12_maths_df.groupby("school_name")["maths_score"].mean()

maths_data_dict = {"Year 9":year_9_maths_series,"Year 10":year_10_maths_series,
                   "Year 11":year_11_maths_series,"Year 12":year_12_maths_series}
maths_score_summary_df = pd.DataFrame(maths_data_dict)
maths_score_summary_df

```

## Reading Score by Year

* Perform the same operations as above for reading scores

```python
year_9_reading_df=school_data_complete.loc[(school_data_complete["year"]==9),:]
year_9_reading_series = year_9_reading_df.groupby("school_name")["reading_score"].mean()
year_10_reading_df=school_data_complete.loc[(school_data_complete["year"]==10),:]
year_10_reading_series = year_10_reading_df.groupby("school_name")["reading_score"].mean()
year_11_reading_df=school_data_complete.loc[(school_data_complete["year"]==11),:]
year_11_reading_series = year_11_reading_df.groupby("school_name")["reading_score"].mean()
year_12_reading_df=school_data_complete.loc[(school_data_complete["year"]==12),:]
year_12_reading_series = year_12_reading_df.groupby("school_name")["reading_score"].mean()

reading_data_dict = {"Year 9":year_9_reading_series,"Year 10":year_10_reading_series,
                   "Year 11":year_11_reading_series,"Year 12":year_12_reading_series}
reading_score_summary_df = pd.DataFrame(reading_data_dict)
reading_score_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0,585,630,645,680]
group_labels = ["<$585","$585-630","$630-645","$645-680"]
spending_ranges = pd.cut(school_summary_df["Per Student Budget"],bins,labels= group_labels,include_lowest = True)
school_summary_df["Spending Ranges (Per Student)"]=spending_ranges
spending_summary_df = school_summary_df.groupby(["Spending Ranges (Per Student)"]).agg({"Average Maths Score":"mean",
                                                                                        "Average Reading Score":"mean",
                                                                                        "% Passing Maths":"mean",
                                                                                        "% Passing Reading":"mean",
                                                                                        "% Overall Passing":"mean"})
#rounding to 2 decimal places
spending_summary_df["Average Maths Score"]=spending_summary_df["Average Maths Score"].round(2)
spending_summary_df["Average Reading Score"]=spending_summary_df["Average Reading Score"].round(2)
spending_summary_df["% Passing Maths"]=spending_summary_df["% Passing Maths"].round(2)
spending_summary_df["% Passing Reading"]=spending_summary_df["% Passing Reading"].round(2)
spending_summary_df["% Overall Passing"]=spending_summary_df["% Overall Passing"].round(2)
spending_summary_df
```
